Geeks for Geeks BFS solution

```python
class Solution:
    # Function to return Breadth First Search Traversal of given graph.
    def bfs(self, adj):
        START = 0
        QUEUE = [START]
        visited = [START]

        while QUEUE:
            CURRENT = QUEUE.pop(0)

            neighbours = adj[CURRENT]
            for neighbour in neighbours:
                if neighbour not in visited:
                    visited.append(neighbour)
                    QUEUE.append(neighbour)
        return visited
```

Comparing dijkstra and bfs

bfs for distance

```python

def bfs(adj, start):
    dist = {start: 0}
    q = [start]

    while q:
        u = q.pop(0)   #(O(n) operation but idc rn)
        #All subsequent elements (from index 1 to the end of the list) must be shifted one position
        #to the left to fill the gap created by the removal.
        #This shifting operation involves moving 'n-1' elements, which takes time proportional to 'n'.


        for v in adj[u]:
            if v not in dist:
                dist[v] = dist[u] + 1
                q.append(v)
    return dist

```

dijkstra for distance

```python
import heapq

def dijkstra(adj, start):
    # CHANGED: set all distances as infinite
    dist = {node: float('inf') for node in adj}
    dist[start] = 0

    # CHANGED: use a priority queue instead of normal queue
    pq = [(0, start)]

    while pq:
        # CHANGED: pop the node with smallest current distance
        d, u = heapq.heappop(pq)

        # ADDED: skip if we already have a better path
        if d > dist[u]:
            continue

        # CHANGED: also get weight weights
        for v, w in adj[u]:
            # ADDED: relaxation
            if dist[u] + w < dist[v]:
                dist[v] = dist[u] + w
                heapq.heappush(pq, (dist[v], v))

    return dist
```

```python
class Solution:

    #Function to return a list containing the DFS traversal of the graph.
    def dfs(self, adj):

        def dfs_util(CURRENT, visited,adj):
            visited.append(CURRENT)
            #neighbour is added right when it is CURRENT
            #since the order in which we discover them is not the order in which we visit them
            neighbours = adj[CURRENT]
            for neighbour in neighbours:
                if neighbour not in visited:
                    dfs_util(neighbour,visited,adj)

            return visited

        START = 0
        visited = []

        return dfs_util(START, visited, adj)
        # code here
```

alternate solution with stack (drawing parallels with bfs queue)


```python
class Solution:

    #Function to return a list containing the DFS traversal of the graph.
    def dfs(self, adj):
        START = 0
        STACK = [START]
        visited = []

        while STACK:
            CURRENT = STACK.pop()  #stack behaviour instead of deque/list
            if CURRENT not in visited:
                visited.append(CURRENT)
                #neighbour is added right when it is CURRENT
                #since the order in which we discover them is not the order in which we visit them

                neighbours = reversed(adj[CURRENT])
                #push neighbors in reverse order so the smallest neighbout gets popped first and visited.
                #(thats what gfg wants)
                for neighbour in neighbours:
                    if neighbour not in visited:
                        STACK.append(neighbour)

        return visited
        # code here
```
